# Visualisation: Earnings Based on College Majors in America

We'll be working on a dataset that was originally released by American Community Survey and holds information about job outcomes of students who graduated from American college between 2010 and 2012. 

Aim of this project is to visualise this dataset to gain intuitive understandings about relationships between college majors, their students, their earnings and etc.

The attributes in this data set and what they represent are as follows.

* Rank - Rank by median earnings (the dataset is ordered by this column).
* Major_code - Major code.
* Major - Major description.
* Major_category - Category of major.
* Total - Total number of people with major.
* Sample_size - Sample size (unweighted) of full-time.
* Men - Male graduates.
* Women - Female graduates.
* ShareWomen - Women as share of total.
* Employed - Number employed.
* Median - Median salary of full-time, year-round workers.
* Low_wage_jobs - Number in low-wage service jobs.
* Full_time - Number employed 35 hours or more.
* Part_time - Number employed less than 35 hours.

Let's start exploring our dataset.

There are 173 original entries in our dataset. We'll see if we can clean the dataset by removing entries with missing values

```python
recent_grads = recent_grads.dropna(axis = 0)
cleaned_data_count = recent_grads.shape[0]
print(cleaned_data_count)
```

There are now 172 entries in our dataset which means there was only 1 row with missing values.

Let's now have a look at various scatter plots between our attributes.

```python
ax_sm = recent_grads.plot(x = "Sample_size", y = "Median", kind = "scatter")
ax_sm.set_title("Median vs Sample Size")
```

Do students in more popular majors make money? Our data seems to say No; in fact, there seems to be a slight negative correlation between industries' median income and sample size, meaning as the sample size gets larger, median income decreases.

```python
ax_su = recent_grads.plot(x = "Sample_size", y = "Unemployment_rate", kind = "scatter")
ax_su.set_title("Unemployment Rate vs Sample Size")
```

```python
ax_fm = recent_grads.plot(x = "Full_time", y = "Median", kind = "scatter")
ax_fm.set_title("Median vs Full Time")
```

Is there any link between the number of full-time employees and median salary? It is hard to say because most data points are distributed along 0 full time workers and the variability amongst is large. It is arguable that there is slight negative trend between number of full time workers and their median income. 

```python
ax_swu = recent_grads.plot(x = "ShareWomen", y = "Unemployment_rate", kind = "scatter")
ax_swu.set_title("Unemployment Rate vs Share Women")
```

```python
ax_mm = recent_grads.plot(x = "Men", y = "Median", kind = "scatter")
ax_mm.set_title("Median vs Men")
```

```python
ax_wm = recent_grads.plot(x = "Women", y = "Median", kind = "scatter")
ax_wm.set_title("Median vs Women")
```

```python
ax_swur = recent_grads.plot(x = "ShareWomen", y = "Median", kind = "scatter")
ax_swur.set_title("Median vs Share Women")
```

Do students that majored in subjects that were majority female make more money? No. This scatter plot of meidan vs share women shows negative trendline between them, meaning industries' median income decreases as the percentage of women workers rises.
